'append': If the table exists, new data will be appended to the existing table.<br>
index=False: This argument specifies that the DataFrame index should not be written as a separate column in the database table.<br>

method="multi": This argument specifies how to insert the data. The "multi" option uses multiple INSERT statements in a single transaction, which can be faster for large datasets.

### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
sql SELECT DISTINCT LAUNCH_SITE FROM SPACEXTBL ORDER BY 1;
```

Explanation:

SELECT DISTINCT LAUNCH_SITE:

SELECT: This is the SQL command used to retrieve data from a database.<br>
DISTINCT: This keyword ensures that the result set contains only unique values. It eliminates duplicate rows.<br>
LAUNCH_SITE: This specifies the column from which we want to retrieve the unique values. In this case, we are interested in the LAUNCH_SITE column.<br>

FROM SPACEXTBL:<br>
This specifies the table from which to retrieve the data. Here, SPACEXTBL is the table containing the launch data.

ORDER BY 1:<br>
ORDER BY: This clause is used to sort the result set.<br>
1: This refers to the first column in the SELECT statement. In this case, it is the LAUNCH_SITE column.<br>
The result set will be sorted in ascending order by the LAUNCH_SITE column by default.<br>



### Task 2


#####  Display 5 records where launch sites begin with the string 'CCA' 


```python
sql SELECT * FROM SPACEXTBL WHERE LAUNCH_SITE LIKE 'CCA%' LIMIT 5;
```

LIKE 'CCA%': This condition uses the LIKE operator, which is used for pattern matching in SQL.<br>
'CCA%': This pattern specifies that the LAUNCH_SITE must start with the letters CCA. The % wildcard character matches any sequence of characters <br>(including an empty sequence).<br>
For example, it will match values like CCAFS SLC 40, CCAFS LC 39A, etc.<br>

LIMIT 5:<br>
This clause specifies that the query should return only the first 5 rows that match the WHERE condition.<br>
This is useful for limiting the number of results returned by the query, especially when dealing with large datasets.<br>


### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
sql SELECT SUM(PAYLOAD_MASS__KG_) AS TOTAL_PAYLOAD FROM SPACEXTBL WHERE PAYLOAD  LIKE '%CRS%';
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
sql SELECT AVG(PAYLOAD_MASS__KG_) AS AVG_PAYLOAD FROM SPACEXTBL WHERE BOOSTER_VERSION = 'F9 v1.1';
```

### Task 5

##### List the date when the first succesful landing outcome in ground pad was acheived.


_Hint:Use min function_ 


```python
sql SELECT MIN(DATE) AS FIRST_SUCCESS_GP FROM SPACEXTBL WHERE Landing_Outcome = 'Success (ground pad)';
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
sql SELECT DISTINCT BOOSTER_VERSION FROM SPACEXTBL WHERE PAYLOAD_MASS__KG_ BETWEEN 4000 AND 6000 AND Landing_Outcome = 'Success (drone ship)';
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python
sql SELECT MISSION_OUTCOME, COUNT(*) AS QTY FROM SPACEXTBL GROUP BY MISSION_OUTCOME ORDER BY MISSION_OUTCOME;
```

Explanation:

SELECT MISSION_OUTCOME, COUNT(*) AS QTY:<br>
SELECT: This is the SQL command used to retrieve data from a database.<br>
MISSION_OUTCOME: This specifies the column we want to include in the result set.<br>
COUNT(*): This function counts the number of rows for each unique value in the MISSION_OUTCOME column.<br>
AS QTY: This renames the COUNT(*) column to QTY in the result set for better readability.<br>

FROM SPACEXTBL:<br>
This specifies the table from which to retrieve the data. Here, SPACEXTBL is the table containing the launch data.<br>

GROUP BY MISSION_OUTCOME:<br>
This clause groups the rows in the SPACEXTBL table by the unique values in the MISSION_OUTCOME column.<br>
For each unique value in the MISSION_OUTCOME column, the query will count the number of rows and include this count in the result set.<br>

ORDER BY MISSION_OUTCOME:<br>
This clause sorts the result set by the MISSION_OUTCOME column in ascending order.<br>
Sorting helps organize the results in a meaningful way, making it easier to analyze the data.<br>


### Task 8



##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
sql SELECT DISTINCT Booster_Version FROM SPACEXTBL WHERE PAYLOAD_MASS__KG_ = (SELECT MAX(PAYLOAD_MASS__KG_) FROM SPACEXTBL) ORDER BY BOOSTER_VERSION;
```

Explanation:

SELECT DISTINCT Booster_Version:<br>
SELECT: This command is used to retrieve data from the database.<br>
DISTINCT: This keyword ensures that the result set will contain only unique values for the Booster_Version column.<br>
Booster_Version: This specifies the column we want to retrieve and ensure uniqueness for.<br>

FROM SPACEXTBL:<br>
This specifies the table from which to retrieve the data. In this case, the table is SPACEXTBL.<br>
WHERE PAYLOAD_MASS__KG_ = (SELECT MAX(PAYLOAD_MASS__KG_) FROM SPACEXTBL):<br>

WHERE: This clause filters the rows in the SPACEXTBL table.<br>
PAYLOAD_MASS__KG_: This specifies the column we are filtering on.<br>
=: This is the equality operator, which is used to compare values.<br>
(SELECT MAX(PAYLOAD_MASS__KG_) FROM SPACEXTBL): This is a subquery that finds the maximum value in the PAYLOAD_MASS__KG_ column from the SPACEXTBL <br>table. The main query will then compare each row's PAYLOAD_MASS__KG_ to this maximum value, effectively filtering the rows to include only those <br>with the maximum payload mass.<br>

ORDER BY BOOSTER_VERSION:<br>
This clause sorts the result set by the Booster_Version column in ascending order.<br>


### Task 9


##### List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015.

**Note: SQLLite does not support monthnames. So you need to use  substr(Date, 6,2) as month to get the months and substr(Date,0,5)='2015' for year.**


```python
%%sql
SELECT 
    substr(Date, 6, 2) AS month, 
    date, 
    booster_version, 
    launch_site 
FROM 
    SPACEXTBL 
WHERE 
    Landing_Outcome = 'Failure (drone ship)' 
    AND substr(Date, 1, 4) = '2015';
```

Explanation:

%%sql
This line is a magic command used in Jupyter Notebooks to indicate that the following lines are SQL code. It is used in conjunction with an SQL extension that allows SQL queries to be run directly in the notebook.


SELECT <br>
    substr(Date, 6, 2) AS month, <br>
    date, <br>
    booster_version, <br>
    launch_site <br>

SELECT: This command is used to specify the columns to retrieve from the table.

substr(Date, 6, 2) AS month: This function extracts a substring from the Date column, starting at the 6th character and taking 2 characters (which <br>corresponds to the month part of the date). The extracted substring is aliased as month.

date: This column is selected as is.<br>
booster_version: This column is selected as is.<br>
launch_site: This column is selected as is.<br>

FROM <br>
    SPACEXTBL <br>

FROM SPACEXTBL: This specifies the table (SPACEXTBL) from which to retrieve the data.<br>

WHERE <br>
    Landing_Outcome = 'Failure (drone ship)' <br>
    AND substr(Date, 1, 4) = '2015';<br>

WHERE: This clause is used to filter the rows returned by the query based on certain conditions.<br>
Landing_Outcome = 'Failure (drone ship)': This condition filters the rows to include only those where the Landing_Outcome column has the value <br>'Failure (drone ship)'.

AND substr(Date, 1, 4) = '2015': This condition further filters the rows to include only those where the first 4 characters of the Date column <br>(which corresponds to the year part of the date) are '2015'.
<br>

Summary:

The query retrieves the month, date, booster version, and launch site for all records in the SPACEXTBL table where the landing outcome was a <br>failure on a drone ship and the launch occurred in the year 2015. The substr function is used to extract parts of the date to filter by year and to <br>display the month separately.<br>








### Task 10




##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order.


```python
sql SELECT Landing_Outcome, COUNT(*) AS QTY FROM SPACEXTBL WHERE DATE BETWEEN '2010-06-04' AND '2017-03-20' GROUP BY Landing_Outcome ORDER BY QTY DESC;
```
